````python
chain_of_thought_prompt_en = '''
Please classify the following review in the {domain} domain as postive, negative or neutral. Please only return the sentiment polarity without explaination.

Your output should be consistent and follow this format:
Input: [user input]
Thought: List all text has emotion in the sentence
Follow output: [list of text has emotion. Each text separated with "|" symbol]
Thought: List all sentiment polarity in each text
Follow output: [list of sentiment polarity in each text. Each sentiment polarity separated with "|" symbol]
Final ouput: [Sentiment polarity]

For instance, if the user give you a sentence "Tôi thích cái dáng váy này nhưng màu sắc không được đẹp cho lắm", your output should be:
Input: Tôi thích cái dáng váy này nhưng màu sắc không được đẹp cho lắm
Thought: List all text has emotion in the sentence
Follow output: Thích cái dáng váy | Màu sắc không được đẹp
Thought: List all sentiment polarity in each text
Follow output: Positive | Negative
Final output: Neutral
'''

chain_of_thought_prompt_vn = '''
"Vui lòng phân loại đánh giá sau đây trong lĩnh vực {domain} là tích cực, tiêu cực hoặc trung lập. Vui lòng chỉ trả lại cường độ tình cảm mà không giải thích.

Kết quả của bạn nên đồng nhất và tuân theo định dạng này:
Input: [đầu vào của người dùng]
Suy nghĩ: Liệt kê tất cả văn bản có cảm xúc trong câu
Kết quả tiếp theo: [danh sách tất cả văn bản có cảm xúc. Mỗi đoạn văn được phân tách bằng ký tự "|" ]
Suy nghĩ: Liệt kê tất cả cường độ tình cảm trong mỗi đoạn văn
Kết quả tiếp theo: [danh sách tất cả cường độ tình cảm trong văn bản. Mỗi cường độ tình cảm được phân tách bằng ký tự "|" ]
Kết quả cuối cùng: [Cường độ tình cảm]"

For example, if the user provides the sentence "Tôi thích cái dáng váy này nhưng màu sắc không được đẹp cho lắm," your output should be:

Input: Tôi thích cái dáng váy này nhưng màu sắc không được đẹp cho lắm
Suy nghĩ: Liệt kê tất cả văn bản có cảm xúc trong câu
Kết quả tiếp theo: Thích cái dáng váy | Màu sắc không được đẹp
Suy nghĩ: Liệt kê tất cả cường độ tình cảm trong mỗi đoạn văn
Kết quả tiếp theo: Tích cực | Tiêu cực
Kết quả cuối cùng: Trung lập
'''

sim_instruction_prompt_en = '''
Please classify the following review in the {domain} domain as postive, negative or neutral. Please only return the sentiment polarity without explaination.

Your output should be consistent and follow this format:
- Output: Sentiment polarity

For instance, if the comment displays a positive sentiment, your output should be:
- Output: Positive
'''

sim_instruction_prompt_vn = '''
Vui lòng phân loại bình luận sau trong lĩnh vực {domain} là tích cực, tiêu cực hoặc trung lập. Vui lòng chỉ trả lại cường độ tâm trạng mà không giải thích.

Đầu ra của bạn nên nhất quán và tuân theo định dạng này:
Đầu ra: Cường độ tâm trạng

Ví dụ, nếu bình luận thể hiện tâm trạng tích cực, đầu ra của bạn nên là:
Đầu ra: Tích cực
'''

com_instruction_prompt_en = '''
As an Sentiment Analysis Model, your task is to analyze the emotional tone of specific user comments. The analysis should result in classifying each comment into one of three predefined categories: Positive, Negative, or Neutral.

Here are the definitions for each category:
1. Positive: The user comment conveys a positive sentiment or favorable opinion about the subject.
2. Negative: The user comment expresses a negative sentiment or unfavorable opinion about the subject.
3. Neutral: The user comment displays neither positive nor negative sentiment, indicating an impartial or balanced perspective.

Please adhere to the following guidelines to ensure accurate analysis:
1. Initially, classify all comments as Neutral until the review indicates otherwise.
2. The emotional tone should be identified based on the given comment alone, without considering previous comments.
3. Exclude elements relating to service delivery or online transactional experiences from the emotional analysis.

Your output should be consistent and follow this format:
- Output: Sentiment polarity

For instance, if the comment displays a positive sentiment, your output should be:
- Output: Positive

By following these guidelines, you'll be able to provide a precise and consistent classification of user comments based on their emotional tone.
'''

com_instruction_prompt_vn = '''
Dưới vai trò một Mô hình Phân tích Tâm trạng, nhiệm vụ của bạn là phân tích tâm trạng cảm xúc của các bình luận cụ thể của người dùng. Phân tích này sẽ dẫn đến việc phân loại mỗi bình luận vào một trong ba danh mục đã định trước: Tích cực, Tiêu cực hoặc Trung lập.

Dưới đây là các định nghĩa cho mỗi danh mục:

Tích cực: Bình luận của người dùng truyền đạt tâm trạng tích cực hoặc ý kiến ưa thích về đề tài.
Tiêu cực: Bình luận của người dùng thể hiện tâm trạng tiêu cực hoặc ý kiến không ưa thích về đề tài.
Trung lập: Bình luận của người dùng không thể hiện tâm trạng tích cực hoặc tiêu cực, cho thấy góc nhìn không thiên vị hoặc cân nhắc.

Vui lòng tuân theo các hướng dẫn sau để đảm bảo phân tích chính xác:
1. Ban đầu, phân loại tất cả bình luận là Trung lập cho đến khi xem xét cho thấy điều ngược lại.
2. Tâm trạng cảm xúc nên được xác định dựa trên bình luận cụ thể đã cho, mà không xem xét các bình luận trước đó.
3. Loại trừ các yếu tố liên quan đến dịch vụ hoặc trải nghiệm giao dịch trực tuyến khỏi phân tích cảm xúc.

Kết quả của bạn nên nhất quán và tuân theo định dạng sau đây:
- Kết quả: Tâm trạng cảm xúc

Ví dụ, nếu bình luận thể hiện tâm trạng tích cực, kết quả của bạn nên là:
- Kết quả: Tích cực

Bằng cách tuân theo các hướng dẫn này, bạn sẽ có khả năng cung cấp một phân loại chính xác và nhất quán của các bình luận của người dùng dựa trên tâm trạng cảm xúc của họ.
'''

input_sentence_en = '''
Input: {sent}
'''

input_sentence_vn = '''
Đầu vào: {sent}
'''
````

function sendQuestions(questions, maxDelayInSeconds) {
    var textareaElement = document.querySelector('#prompt-textarea');
    var sendButtonElement = document.querySelector('[data-testid="send-button"]');
    
    questions.forEach(function(question, index) {
        var randomDelay = Math.random() * maxDelayInSeconds * 1000; // Convert to milliseconds
        setTimeout(function () {
            textareaElement.value = question;

            var inputEvent = new Event('input', {
                bubbles: true,
                cancelable: true
            });

            textareaElement.dispatchEvent(inputEvent);

            sendButtonElement.click();
        }, randomDelay);
    });
};

// Example usage
with a maximum delay of 5 seconds (5000 milliseconds):

var questionsToSend = ["Question 1", "Question 2", "Question 3"];
sendQuestions(questionsToSend, 2);
